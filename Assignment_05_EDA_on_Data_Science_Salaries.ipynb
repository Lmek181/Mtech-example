```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/ajaverett/data/refs/heads/main/data_analytics_salaries.csv")

df['salary_in_usd'].mean()
```

```python
entry=df[(df['experience_level']=='EN') &(df['employment_type']=='FT')]
entry['salary_in_usd'].mean()
```

```python
df[['experience_level', 'salary']].groupby('experience_level').mean().sort_values(by='salary')
```

```python
remote=df[['remote_ratio','experience_level']].groupby('experience_level').mean().sort_values(by='remote_ratio')
remote
```

```python
increase_remote = (df
 [df['work_year']==2022]
  .filter(items=['remote_ratio','experience_level'])
  .groupby('experience_level')
  .mean()
  .reset_index()
  .sort_values(by='remote_ratio')
)
increase_remote
```

```python
US_job = (df
    [df['company_location']=='US']
    .filter(items=['job_title','company_location'])
    .groupby('job_title')
    .agg({'company_location':'sum'})
    .reset_index()
    .sort_values(by='job_title',ascending=True)
)
US_job
```

```python
job_US=df[(df['company_location']=='US') &(df['job_title']=='Data Scientist')]
```

```python
median_salary = (df
 .filter(items=['job_title', 'salary_in_usd'])
  .groupby('job_title')
  .median()
  .reset_index()
  .sort_values(by='job_title',ascending=True)
)
median_salary
```

```python
(df
 .filter(items=['job_title', 'salary_in_usd'])
  .groupby('job_title')
  .std()
  .reset_index()
  .sort_values(by='job_title',ascending=True)
)
```

```python
(df
 .filter(items=['job_title', 'salary_in_usd'])
  .groupby('job_title')
  .mean()
  .reset_index()
  .sort_values(by='job_title',ascending=True)
)
```

```python
(df
 .groupby('work_year')
 .agg({'salary':'mean'})
)
```

```python
(df
 .filter(items=['work_year'==2021, 'salary','job_title'])
  .groupby('salary')
  .mean()
  .reset_index()
  .sort_values(by='work_year',ascending=True)
)
```
